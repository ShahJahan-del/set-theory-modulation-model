```python
import json
import torch
import torch.nn as nn
import numpy as np
import random

# 1. CHARGEMENT DU DATASET PROPRE
with open("datasets/variations_3000_impressionist_v2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

random.seed(42)
random.shuffle(data)

# 2. CRÉATION DES VOCABULAIRES UNIQUES POUR LE ONE-HOT
sets_vocab = sorted(list(set(r["Start_Set (SS)"].strip() for r in data)))
transpos_vocab = sorted(list(set(r["Start_Transposition (ST)"].strip() for r in data)))
modes_vocab = sorted(list(set(r["Start_Mode (SM)"].strip() for r in data)))

# Blocs d'arrivée uniques (les classes à prédire)
Y_strings = [f"{r['Target_Set (TS)'].strip()}||{r['Target_Transposition (TT)'].strip()}||{r['Target_Mode (TM)'].strip()}" for r in data]
vocab_arrivee = {val: idx for idx, val in enumerate(sorted(list(set(Y_strings))))}
idx_to_arrivee = {v: k for k, v in vocab_arrivee.items()}

# Fonctions d'encodage One-Hot
def to_one_hot(val, vocab):
    vec = [0] * len(vocab)
    if val in vocab:
        vec[vocab.index(val)] = 1
    return vec

# 3. PRÉPARATION DES MATRICES
X_list, Y_list = [], []
for i, row in enumerate(data):
    try:
        # Encodage de chaque caractéristique textuelle en commutateurs 0/1
        vec_set = to_one_hot(row["Start_Set (SS)"].strip(), sets_vocab)
        vec_transpo = to_one_hot(row["Start_Transposition (ST)"].strip(), transpos_vocab)
        vec_mode = to_one_hot(row["Start_Mode (SM)"].strip(), modes_vocab)
        pivots_norm = float(row["Common_Tones (CT)"]) / 7.0
        
        # L'entrée est la fusion de tous les One-Hot + la contrainte
        X_list.append(vec_set + vec_transpo + vec_mode + [pivots_norm])
        Y_list.append(vocab_arrivee[Y_strings[i]])
    except KeyError:
        continue

X_all = torch.tensor(X_list, dtype=torch.float32)
Y_all = torch.tensor(Y_list, dtype=torch.long)

# Split Train (2500) / Test (500)
X_train, X_test = X_all[:2500], X_all[2500:]
Y_train, Y_test = Y_all[:2500], Y_all[2500:]

# 4. RÉSEAU DE CLASSIFICATION LARGE
class CategoricalHarmonicClassifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(CategoricalHarmonicClassifier, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3), # Évite le surapprentissage
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, num_classes)
        )
    def forward(self, x): 
        return self.net(x)

# L'entrée fait maintenant la somme des tailles des vocabulaires + 1
input_dimension = len(sets_vocab) + len(transpos_vocab) + len(modes_vocab) + 1
model = CategoricalHarmonicClassifier(input_dim=input_dimension, num_classes=len(vocab_arrivee))

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Entraînement
model.train()
for epoch in range(300):
    optimizer.zero_grad()
    loss = criterion(model(X_train), Y_train)
    loss.backward()
    optimizer.step()

# 5. TEST DE GÉNÉRALISATION EN MODE PROBABILISTE (TOP 5)
model.eval()
with torch.no_grad():
    outputs = model(X_test)
    
    # On extrait les 5 meilleures prédictions pour chaque ligne de test
    _, top5_preds = torch.topk(outputs, k=5, dim=1)
    
    # On regarde si la vraie cible est dans ces 5 propositions
    reussites_top5 = 0
    for i in range(len(Y_test)):
        if Y_test[i] in top5_preds[i]:
            reussites_top5 += 1
            
    accuracy_top5 = (reussites_top5 / len(Y_test)) * 100
    print(f"🎯 Précision Top 5 sur les blocs : {accuracy_top5:.1f}%")
```

```python
# 6. TON GÉNÉRATEUR DE MODULATIONS (L'ORACLE)
def suggerer_modulations(set_dep, transpo_dep, mode_dep, pivots=4):
    model.eval()
    with torch.no_grad():
        try:
            vec_set = to_one_hot(set_dep, sets_vocab)
            vec_transpo = to_one_hot(transpo_dep, transpos_vocab)
            vec_mode = to_one_hot(mode_dep, modes_vocab)
            pivots_norm = float(pivots) / 7.0
            
            in_vector = torch.tensor([vec_set + vec_transpo + vec_mode + [pivots_norm]], dtype=torch.float32)
            logits = model(in_vector)
            probabilities = torch.softmax(logits, dim=1).numpy()[0]
            
            best_indices = np.argsort(probabilities)[::-1][:5]
            
            print(f"\n🔮 [Assistant] Depuis {set_dep} {transpo_dep} {mode_dep} ({pivots} pivots) :")
            for i, idx in enumerate(best_indices):
                print(f"  Option {i+1} ({probabilities[idx]*100:.1f}%) -> {idx_to_arrivee[idx]}")
        except ValueError:
            print("Erreur : Un des termes n'est pas reconnu dans le vocabulaire.")
```

```python
# Exemple d'utilisation pour tes morceaux :
suggerer_modulations("Diatonic", "C", "Mixolydian", pivots=4)
```

```python
def generer_framework_sur_mesure(set_dep, transpo_dep, mode_dep, plan_pivots=[4, 3, 5, 2]):
    """
    Génère un voyage harmonique complet basé sur un plan de pivots personnalisé.
    Permet de contrôler la tension harmonique du morceau étape par étape.
    """
    courant_set = set_dep
    courant_transpo = transpo_dep
    courant_mode = mode_dep
    
    voyage = [f"{courant_set} {courant_transpo} {courant_mode}"]
    
    print(f"🎼 --- GÉNÉRATION DU FRAMEWORK HARMONIQUE ({len(plan_pivots)} mouvements) ---")
    print(f"Départ : {voyage[0]}")
    print(f"Plan de tension programmé (pivots) : {plan_pivots}\n")
    
    for pas, pivots_etape in enumerate(plan_pivots):
        model.eval()
        with torch.no_grad():
            # 1. Encodage du bloc courant
            vec_set = to_one_hot(courant_set, sets_vocab)
            vec_transpo = to_one_hot(courant_transpo, transpos_vocab)
            vec_mode = to_one_hot(courant_mode, modes_vocab)
            
            # 2. On injecte la contrainte spécifique à CETTE étape
            pivots_norm = float(pivots_etape) / 7.0
            
            in_vector = torch.tensor([vec_set + vec_transpo + vec_mode + [pivots_norm]], dtype=torch.float32)
            
            # 3. Prédiction et échantillonnage probabiliste (Top 5)
            logits = model(in_vector)
            probabilities = torch.softmax(logits, dim=1).numpy()[0]
            
            top5_indices = np.argsort(probabilities)[::-1][:5]
            top5_probs = probabilities[top5_indices]
            top5_probs /= top5_probs.sum() 
            
            idx_choisi = np.random.choice(top5_indices, p=top5_probs)
            bloc_suivant = idx_to_arrivee[idx_choisi]
            
            # 4. Extraction pour l'étape suivante
            next_set, next_transpo, next_mode = bloc_suivant.split("||")
            
            # Stockage avec un joli formatage pour voir la contrainte appliquée
            voyage.append(f" └──► Exige [{pivots_etape} pivots] ➔ {next_set} {next_transpo} {next_mode}")
            
            # Remplacement des variables pour le prochain saut
            courant_set = next_set
            courant_transpo = next_transpo
            courant_mode = next_mode

    # Affichage du résultat final
    for etape in voyage[1:]:
        print(etape)
    print("-------------------------------------------------------------------\n")
```
